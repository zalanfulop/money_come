```python
import pandas as pd
```

```python
# Load data
df_transacts = pd.read_csv("2025_07_transactions.csv")

# Convert the 'date' column to datetime
df_transacts["date"] = pd.to_datetime(df_transacts["date"])

# Convert all object columns (except date) to string dtype
object_cols = df_transacts.select_dtypes(include="object").columns
df_transacts[object_cols] = df_transacts[object_cols].astype("string")

# Set 'date' as index and drop the old index
#df_transacts = df_transacts.set_index("date")
```

```python
# Create su
# Get unique values from specific columns, sort alphabetically
vendors = df_transacts["vendor"].dropna().unique()
vendors = sorted(vendors)
dates = df_transacts["date"].dropna().unique()
dates = sorted(dates)
product_types = df_transacts["product_type"].dropna().unique()
product_types = sorted(product_types)
product_names = df_transacts["product_name"].dropna().unique()
product_names = sorted(product_names)
pay_methods = df_transacts["pay_method"].dropna().unique()
pay_methods = sorted(pay_methods)

# Convert to DataFrame for CSV export
vendors_df = pd.DataFrame(vendors, columns=["vendor"])
dates_df = pd.DataFrame(dates, columns=["date"])
product_types_df = pd.DataFrame(product_types, columns=["product_type"])
product_names_df = pd.DataFrame(product_names, columns=["product_name"])
pay_methods_df = pd.DataFrame(pay_methods, columns=["pay_method"])


# Save to CSV
vendors_df.to_csv("vendors.csv", index=False)
dates_df.to_csv("dates.csv", index=False)
product_types_df.to_csv("product_types.csv", index=False)
product_names_df.to_csv("product_names.csv", index=False)
pay_methods_df.to_csv("pay_methods.csv", index=False)
```
